```python
metadata_df = pd.read_csv("C://Users//Admin//Documents//medical-research-assistant//data//raw//metadata.csv", low_memory=False)

```

```python
metadata_df.info()
```

```python
metadata_df = metadata_df[['sha', 'publish_time', 'journal', 'doi', 'license', 'url']]
metadata_df = metadata_df.dropna(subset=['sha'])
```

```python
metadata_df.info()
```

```python
merged_df = df1.merge(
    metadata_df,
    how='left',
    left_on='paper_id',
    right_on='sha'
)

# Optional: drop the duplicate 'sha'
merged_df.drop(columns=['sha'], inplace=True)
```

```python
merged_df.info()
```

```python
# Save everything — no filtering
merged_df.to_parquet("C://Users//Admin//Documents//medical-research-assistant//data//processed//cord19_enriched.parquet", index=False)

print("✅ Saved all 100K enriched records to cord19_enriched.parquet")

```
